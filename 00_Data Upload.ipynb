```python
params = {
    'host': 'localhost',  #connecting to local version of psql
    'user': 'agar',
    'dbname': 'properties',    # DB that we are connecting to (initially 'votes' but changed to 'properties')
    'port': 5432          # port we opened on AWS
}

# unpack params dict to connect
connection = pg.connect(**params)
```

```python
# Connect and create database, disconnect, and reconnect to the right database
connection = connect(**params, 'dbname': 'votes')
connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)
connection.cursor().execute('CREATE DATABASE properties;')
connection.close()
```

```python
# Read in Assessors Parces Data
properties_df = pd.read_csv("Assessor_Parcels_Data_-_2019.csv", low_memory=False)
properties_df.head()
```

```python
# Create Assessors Table in psql
# This table has Preperty Location, General Use Type, Existing Structure Areas, and AIN ID Numbers
connection_string = f'postgres://agar:{params["host"]}@{params["host"]}:{params["port"]}/properties'
engine = create_engine(connection_string, pool_pre_ping=True)

properties_df.iloc[:0].to_sql("Assessors_2019", engine, index=False)
properties_df.iloc[:].to_sql("Assessors_2019", engine, index=False, if_exists = 'append', chunksize = 1000)
```

```python
# Read in Parcels Data
# This table has Preperty Location, General Use Type, Existing Structure Areas, Lot Area, and AIN ID Numbers
parcels_df = pd.read_csv("Parcels.csv", low_memory=False)
parcels_df.head()
```

```python
# Create Parcels Table in psql
# This table has Preperty Location, General Use Type, Existing Structure Areas, Lot Area, and AIN ID Numbers
parcels_df.iloc[:0].to_sql("parcels", engine, index=False)
parcels_df.iloc[:].to_sql("parcels", engine, index=False, if_exists = 'append', chunksize = 1000)
```

```python
#Test Query from Parcels 

query = """
SELECT * FROM parcels
WHERE "ShapeSTArea" >= 5000
LIMIT 3;
"""
df_parcels = pd.read_sql_query(query, connection)
df_parcels
```
